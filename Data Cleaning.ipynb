# 2.Data Cleaning

## 2.1. Remove Percentage (%) and convert to float datatype

```python
# Remove "%" and convert to float for the "Agricultural Land (%)" column
world_data_df["Agricultural Land( %)"] = world_data_df["Agricultural Land( %)"].str.replace("%", "").astype(float)

world_data_df["Out of pocket health expenditure"] = world_data_df["Out of pocket health expenditure"].str.replace("%", "").astype(float)
world_data_df["Population: Labor force participation (%)"] = world_data_df["Population: Labor force participation (%)"].str.replace("%", "").astype(float)
world_data_df["Tax revenue (%)"] = world_data_df["Tax revenue (%)"].str.replace("%", "").astype(float)
world_data_df["Total tax rate"] = world_data_df["Total tax rate"].str.replace("%", "").astype(float)
world_data_df["Unemployment rate"] = world_data_df["Unemployment rate"].str.replace("%", "").astype(float)
world_data_df["CPI Change (%)"] = world_data_df["CPI Change (%)"].str.replace("%", "").astype(float)
world_data_df["Forested Area (%)"] = world_data_df["Forested Area (%)"].str.replace("%", "").astype(float)
world_data_df["Gross primary education enrollment (%)"] = world_data_df["Gross primary education enrollment (%)"].str.replace("%", "").astype(float)
world_data_df["Gross tertiary education enrollment (%)"] = world_data_df["Gross tertiary education enrollment (%)"].str.replace("%", "").astype(float)
```

## 2.2.  Remove Comma (,) and convert to float datatype

```python
# Remove commas and convert the "Density (P/Km2)" column from string to float
world_data_df["Density\n(P/Km2)"] = world_data_df["Density\n(P/Km2)"].str.replace(",", "").astype(float)

world_data_df["Land Area(Km2)"] = world_data_df["Land Area(Km2)"].str.replace(",", "").astype(float)
world_data_df["Armed Forces size"] = world_data_df["Armed Forces size"].str.replace(",", "").astype(float)
world_data_df["Co2-Emissions"] = world_data_df["Co2-Emissions"].str.replace(",", "").astype(float)
world_data_df["Population"] = world_data_df["Population"].str.replace(",", "").astype(float)
world_data_df["Urban_population"] = world_data_df["Urban_population"].str.replace(",", "").astype(float)
world_data_df["CPI"] = world_data_df["CPI"].str.replace(",", "").astype(float)
```

## 2.3. Remove Dollar ($) and convert to float datatype

```python
#If comma exists, deal with it first
world_data_df["GDP"] = world_data_df["GDP"].str.replace(",", "")
```

```python
world_data_df["Gasoline Price"] = world_data_df["Gasoline Price"].str.replace("$", "").astype(float)
world_data_df["GDP"] = world_data_df["GDP"].str.replace("$", "").astype(float)
world_data_df["Minimum wage"] = world_data_df["Minimum wage"].str.replace("$", "").astype(float)
```

## 2.4. Drop Columns

```python
# List of columns to drop
columns_to_drop = ['Capital/Major City', 'Currency-Code', 'Largest city', 'Official language']

# Drop the columns from the DataFrame
world_data_df.drop(columns=columns_to_drop, inplace=True)
```

## 2.2 Rename to identify 'Country' as Primary key

```python
# Replace 'Country' where it equals 'Micronesia' with 'Federated States of Micronesia'
HDI_df.loc[HDI_df['Country'] == 'Micronesia', 'Country'] = 'Federated States of Micronesia'

HDI_df.loc[HDI_df['Country'] == 'Palestine', 'Country'] = 'Palestinian National Authority'
HDI_df.loc[HDI_df['Country'] == 'Ireland', 'Country'] = 'Republic of Ireland'
HDI_df.loc[HDI_df['Country'] == 'Bahamas', 'Country'] = 'The Bahamas'
HDI_df.loc[HDI_df['Country'] == 'Gambia', 'Country'] = 'The Gambia'
```

# 7. Feature Engineering

## 7.1. Feature Engineering for Logistic Regression (Binary Classification)

```python
import numpy as np

# Create a new column 'HDI-UN_Binary' based on the conditions in filtered_df2
filtered_df2['HDI-UN_Binary'] = np.where(filtered_df2['HDI-UN'] > 0.5, 'Good Life Quality', 'Poor Life Quality')
```

```python
filtered_df2
```

## 7.2. Feature Engineering for Multiclass Classification

```python
import numpy as np

# Define the conditions and corresponding labels
conditions = [
    (filtered_df2['HDI-UN'] >= 0.76),
    (filtered_df2['HDI-UN'] >= 0.51) & (filtered_df2['HDI-UN'] <= 0.75),
    (filtered_df2['HDI-UN'] >= 0.26) & (filtered_df2['HDI-UN'] <= 0.5),
    (filtered_df2['HDI-UN'] >= 0) & (filtered_df2['HDI-UN'] <= 0.25)
]

labels = ['Very Good Life Quality', 'Good Life Quality', 'Poor Life Quality', 'Very Poor Life Quality']

# Create the 'HDI-UN_Multiple' column based on the conditions
filtered_df2['HDI-UN_Multiple'] = np.select(conditions, labels, default='Unknown')
```
